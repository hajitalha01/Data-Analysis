```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sklearn for Linear Regression and metrics
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Load your ice cream CSV file
data = pd.read_csv('/content/Ice Cream.csv')

# Check first 5 rows
data.head()
```

```python
# X = independent variable (Temperature)
# Y = dependent variable (Revenue)
X = data[['Temperature']]
Y = data['Revenue']
```

```python
# Create Linear Regression model
model = LinearRegression()

# Fit model on data
model.fit(X, Y)

# Get slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (β1): {slope}")
print(f"Intercept (β0): {intercept}")
```

```python
# Predicted Revenue
Y_pred = model.predict(X)

# Add predicted values to dataframe
data['Predicted_Revenue'] = Y_pred
data.head()
```

```python
# Mean Squared Error
mse = mean_squared_error(Y, Y_pred)

# Root Mean Squared Error
rmse = np.sqrt(mse)

# Mean Absolute Error
mae = mean_absolute_error(Y, Y_pred)

# R-squared
r2 = r2_score(Y, Y_pred)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
print(f"R-squared: {r2}")
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X, Y, color='blue', label='Actual Revenue')
plt.plot(X, Y_pred, color='red', label='Predicted Revenue')
plt.xlabel('Temperature (°C)')
plt.ylabel('Revenue')
plt.title('Ice Cream Revenue vs Temperature')
plt.legend()
plt.show()
```
